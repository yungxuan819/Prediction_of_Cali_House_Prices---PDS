```python
import pandas as pd #light package to do multiple array for 2 dimension
import numpy as np #package for data frame
import matplotlib.pyplot as plt #do some visualisation
import seaborn as sns #to create some sort of histogram or heap map
```

```python
new_data = pd.read_csv('New_Cali_Houses.csv') #Load New Dataset
new_data.head() # Display first 5 rows of the dataset
```

```python
#Set House as Index
new_data_process = new_data.set_index('House')
```

```python
#Create dummy variables for categorical variable
new_data_dummy = pd.get_dummies(new_data_process)
new_data_dummy = pd.get_dummies(new_data_process,drop_first=True)
new_data_dummy.head() # Display first 5 rows of the set
```

```python
#Call input variable --must same the column name as save model
new = new_data_dummy[['Median_Income','Tot_Rooms','Tot_Bedrooms','Households','Distance_to_coast']] #create data frame
#input_new = new_data_dummy[['Median_Income','Tot_Rooms','Tot_Bedrooms','Households','Distance_to_coast']]
new.head() # Display first 5 rows of the new data set
```

```python
#Feature Scalling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(new)
input_new = scaler.transform(new)#set the variable after feature scalling
```

```python
input_new # Display the new variable
```

```python
#Load Saved Model
import pickle

filename = 'reg_model_save.sav'

loaded_model = pickle.load(open(filename,'rb')) #rb - read binary same as open model/file
```

```python
#Predict
import statsmodels.api as sm
input_new = sm.add_constant(input_new)
predict_output = loaded_model.predict(input_new)#use the feature scalling variable to predict using save model
predict_output = pd.DataFrame({'Predicted Price':predict_output}) #Create Data Frame
```

```python
predict_output.head() # Display the first 5 rows of the predicted price
```

```python
#Create dataframe

new_predict = pd.concat([new_data,predict_output], axis = 1)
```

```python
new_predict.head() #display the first 5 rows of data
```

```python
new_predict.to_csv('new_predict_pr.csv',index=False) # Saving the DataFrame 'new_predict' into a CSV called 'new_price_pr.csv'
```
